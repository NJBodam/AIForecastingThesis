```python
import pandas as pd

# Replace 'your_file.csv' with the actual path to your CSV file
file_path = '/Users/bodamjerry/Desktop/ucmr docs/UIFeedback2.csv'
df = pd.read_csv(file_path)

# Display the first few rows to understand the data
columns_to_check = ['secondSurvey_question1', 'secondSurvey_question3', 'secondSurvey_question2']
df_cleaned_all = df.dropna(subset=columns_to_check, how='all')

print(df_cleaned_all.head())
```

```python
import string

def preprocess_text(text):
    if isinstance(text, str):  # Ensure it's a string
        text = text.lower()
        text = text.translate(str.maketrans('', '', string.punctuation))
        return text
    return '' # Or some other placeholder for non-string values

for col in ['secondSurvey_question1', 'secondSurvey_question3', 'secondSurvey_question2']:
    df_cleaned_all[col] = df_cleaned_all[col].apply(preprocess_text)

print(df_cleaned_all[['secondSurvey_question1', 'secondSurvey_question3', 'secondSurvey_question2']].head())
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

try:
    sid = SentimentIntensityAnalyzer()
except LookupError:
    nltk.download('vader_lexicon')
    sid = SentimentIntensityAnalyzer()

def get_sentiment(text):
    if isinstance(text, str):
        scores = sid.polarity_scores(text)
        if scores['compound'] >= 0.05:
            return 'Positive'
        elif scores['compound'] <= -0.05:
            return 'Negative'
        else:
            return 'Neutral'
    return 'Neutral'

for col in ['secondSurvey_question1', 'secondSurvey_question3', 'secondSurvey_question2']:
    df[col + '_sentiment'] = df[col].apply(get_sentiment)
    print(f"\nSentiment distribution for {col}:")
    print(df[col + '_sentiment'].value_counts())
```

```python
import xml.etree.ElementTree as ET
import json

def parse_bible_xml(xml_path):
    tree = ET.parse(xml_path)
    root = tree.getroot()

    bible_data = {
        "language": "English",
        "translation": "NKJV",
        "books": []
    }

    for book in root.findall(".//BIBLEBOOK"):
        book_name = book.attrib.get("bname", "Unknown Book")
        book_number = int(book.attrib.get("bnumber", 0))
        chapters = []

        for chapter in book.findall("CHAPTER"):
            chapter_number = int(chapter.attrib.get("cnumber", 0))
            verses = []

            for verse in chapter.findall("VERS"):
                verse_number = int(verse.attrib.get("vnumber", 0))
                verse_text = verse.text or ""
                verses.append({
                    "verse": verse_number,
                    "text": verse_text.strip()
                })

            chapters.append({
                "chapter": chapter_number,
                "verses": verses
            })

        bible_data["books"].append({
            "name": book_name,
            "number": book_number,
            "chapters": chapters
        })

    return bible_data

def write_fsb_format(data, output_path):
    with open(output_path, 'w', encoding='utf-8') as fsb_file:
        json.dump(data, fsb_file, indent=2, ensure_ascii=False)

if __name__ == "__main__":
    xml_input_path = "/Users/bodamjerry/Downloads/NKJV.xml"
    fsb_output_path = "/Users/bodamjerry/Downloads/English_NKJV.fsb"

    print("Parsing XML...")
    bible_content = parse_bible_xml(xml_input_path)

    print("Writing FSB...")
    write_fsb_format(bible_content, fsb_output_path)

    print(f"✅ Done. File saved as {fsb_output_path}")
```

```python
import xml.etree.ElementTree as ET
import json

def convert_xmlb_to_fsb_from_file(xml_file_path, copyright_info=""):
    """Converts an XMLBIBLE XML file to the FSB JSON format.

    Args:
        xml_file_path: The path to the XML Bible file.
        copyright_info: The copyright information.

    Returns:
        A JSON string in the FSB format.
    """
    try:
        tree = ET.parse(xml_file_path)
        root = tree.getroot()
    except FileNotFoundError:
        print(f"Error: XML file not found at {xml_file_path}")
        return None
    except ET.ParseError as e:
        print(f"Error parsing XML file: {e}")
        return None

    bible_name = root.get('biblename', "Unknown")
    books_data = []

    for book_element in root.findall('BIBLEBOOK'):
        book_number = book_element.get('bnumber')
        book_name = book_element.get('bname')

        chapters_data = []
        for chapter_element in book_element.findall('CHAPTER'):
            chapter_number = chapter_element.get('cnumber')
            verses_data = []
            for verse_element in chapter_element.findall('VERS'):
                verse_number = verse_element.get('vnumber')
                verse_value = verse_element.text.strip() if verse_element.text is not None else ""
                verses_data.append({
                    "number": verse_number,
                    "value": verse_value
                })
            chapters_data.append({
                "number": chapter_number,
                "verses": verses_data
            })

        books_data.append({
            "number": book_number,
            "name": book_name,
            "chapters": chapters_data
        })

    fsb_data = [
        "99d99c982f8",
        {
            "name": bible_name,
            "copyright": copyright_info,
            "books": books_data
        }
    ]
    return json.dumps(fsb_data, ensure_ascii=False, indent=None)

if __name__ == "__main__":
    xml_file = "/Users/bodamjerry/Downloads/Bible_English_NCV.xml"  # Replace with the actual name of your XML file
    copyright_information = "" # Add copyright information if available

    fsb_json_output = convert_xmlb_to_fsb_from_file(xml_file, copyright_information)

    if fsb_json_output:
        output_file_path = "/Users/bodamjerry/Downloads/Bible_English_NCV.fsb" # Use the desired output path
        with open(output_file_path, "w", encoding="utf-8") as f:
            f.write(fsb_json_output)
        print(f"Conversion complete. The output has been saved to '{output_file_path}'")
```
